```python
from langchain_openai import OpenAIEmbeddings
```

```python
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
```

```python
embeddings.embed_query("hi buddy how are you? doing?")
```

```python
embeddings.embed_query("hi buddy how are you? doing?") == embeddings.embed_query(
    "hi buddy how are you? doing?"
)
```
